```python
# Sample data for a few days
data = {
    'Date': ['2023-10-20', '2023-10-23', '2023-10-24', '2023-10-25', '2023-10-26'],
    'Open': [120.5, 121.5, 122.0, 121.8, 123.0],
    'High': [121.0, 123.0, 123.5, 122.5, 124.0],
    'Low': [119.8, 120.5, 121.2, 120.8, 122.0],
    'Close': [120.8, 122.5, 122.2, 121.5, 123.5],
    'Volume': [1500, 2200, 1800, 1300, 2100]
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Plotting the candlestick chart
mpf.plot(df, type='candle', style='charles', title='Candlestick Chart', ylabel='Price', volume=True)
```
