```python
import pandas as pd
```

```python
STOCK_LIST =  ['NVDA', 'TSLA', 'AMZN', 'MSFT', 'COST']

stocks_dfs_dict = {}

for symbol in STOCK_LIST:
    file_name = f"{symbol.lower()}_stock.csv"
    path_csv = f"data/{file_name}"
    stocks_dfs_dict[f"{symbol.lower()}"] = pd.read_csv(path_csv,index_col='Date')
```
